```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
%matplotlib inline
```

```python
df = pd.read_excel('houston_pollen.xlsx', sheet_name='2021', parse_dates=['date'])
df.head()
```

```python
plt.figure(figsize=(15,12))
plt.plot(df['month'], df['Pollen Total'], c='black', lw=4, label='Pollen Total')
plt.plot(df['month'], df['Tree Total'], c='b', lw=2, label='Tree Total')
plt.plot(df['month'], df['Grass Total'], c='r', lw=2, label='Grass Total')
plt.plot(df['month'], df['Weed Total'], c='orange', lw=2, label='Weed Total')
plt.xlabel('Month', fontsize=20)
plt.ylabel('Pollen Count', fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(fontsize=20)
plt.title('Houston Pollen Count in 2021', fontsize=20)

ax = plt.gca()
fmt = "{x:,.0f}"
tick = mpl.ticker.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)

plt.savefig('houston_pollen_total_2021_lineplot.png')
```

```python
plt.figure(figsize=(15,12))
plt.bar(df['month'], df['Tree Total'], label='Tree Total', color='darkblue')
plt.bar(df['month'], df['Grass Total'], label='Grass Total', bottom=df['Tree Total'], color='red')
plt.bar(df['month'], df['Weed Total'], label='Weed Total', bottom=df['Tree Total']+df['Grass Total'], color='orange')
plt.legend(fontsize=20)
plt.xlabel('Month', fontsize=20)
plt.ylabel('Pollen Count', fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.title('Houston Pollen Count in 2021', fontsize=20)

ax = plt.gca()
fmt = "{x:,.0f}"
tick = mpl.ticker.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)

plt.savefig('houston_pollen_total_2021_barplot.png')
```

```python
df_total = pd.read_excel('houston_pollen.xlsx', sheet_name='2021_total')

df_total_may = df_total[['pollen', 'May']][df_total['May'] > 0]
df_total_jun = df_total[['pollen', 'Jun']][df_total['Jun'] > 0]
df_total_jul = df_total[['pollen', 'Jul']][df_total['Jul'] > 0]
df_total_aug = df_total[['pollen', 'Aug']][df_total['Aug'] > 0]

colors = ("skyblue", "orange", "beige", "grey", "lightgreen", "indigo", "red", "black")
wp = {'linewidth' : 1, 'edgecolor' : "green"}

fig, ax = plt.subplots(nrows=2, ncols=2, figsize =(12, 12))

wedges, texts, autotexts = ax[0,0].pie(df_total_may['May'], labels = df_total_may['pollen'], autopct='%1.1f%%', 
                                  colors = colors, startangle = 90, wedgeprops = wp,
                                  textprops = dict(color ="magenta", fontsize=15))

ax[0,0].set_title('Contribution of Pollen Count in May', fontsize=15)
plt.setp(autotexts, size = 15, weight ="bold")

wedges, texts, autotexts = ax[0,1].pie(df_total_jun['Jun'], labels = df_total_jun['pollen'], autopct='%1.1f%%', 
                                  colors = colors, startangle = 90, wedgeprops = wp,
                                  textprops = dict(color ="magenta", fontsize=15))

ax[0,1].set_title('Contribution of Pollen Count in June', fontsize=15)
plt.setp(autotexts, size = 15, weight ="bold")

wedges, texts, autotexts = ax[1,0].pie(df_total_jul['Jul'], labels = df_total_jul['pollen'], autopct='%1.1f%%', 
                                  colors = colors, startangle = 90, wedgeprops = wp,
                                  textprops = dict(color ="magenta", fontsize=15))

ax[1,0].set_title('Contribution of Pollen Count in July', fontsize=15)
plt.setp(autotexts, size = 15, weight ="bold")

wedges, texts, autotexts = ax[1,1].pie(df_total_aug['Aug'], labels = df_total_aug['pollen'], autopct='%1.1f%%', 
                                  colors = colors, startangle = 90, wedgeprops = wp,
                                  textprops = dict(color ="magenta", fontsize=15))

ax[1,1].set_title('Contribution of Pollen Count in August', fontsize=15)
plt.setp(autotexts, size = 15, weight ="bold")
plt.tight_layout()

plt.savefig('contribution_pollen_total_may_june_july_august.png')
```
